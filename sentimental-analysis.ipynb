```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import json
import matplotlib.pyplot as plt

# Download the VADER lexicon for sentiment analysis
nltk.download('vader_lexicon')

# Load the JSON data containing video titles
with open('sourav_joshi_data.json', 'r') as file:
    data = json.load(file)

# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Function to perform sentiment analysis on video titles
def analyze_sentiment(title):
    # Get sentiment scores for the title
    scores = sid.polarity_scores(title)
    # Classify sentiment based on compound score
    if scores['compound'] >= 0.05:
        return 'Positive'
    elif scores['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Perform sentiment analysis on each video title
sentiment_counts = {'Positive': 0, 'Negative': 0, 'Neutral': 0}
for video in data:
    title = video['title']
    sentiment = analyze_sentiment(title)
    sentiment_counts[sentiment] += 1

# Plot a pie chart to visualize the distribution of sentiments
labels = sentiment_counts.keys()
sizes = sentiment_counts.values()

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Sentiments in Video Titles (Sourav Joshi Vlogs)')
plt.axis('equal')
plt.show()
```
