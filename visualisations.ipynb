Comparing numner of views in the month of February 2024

```python
import json
from datetime import datetime
import matplotlib.pyplot as plt

# Function to extract views for February from a JSON file
def extract_views_for_february(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    views_by_day = {}
    for video in data:
        published_date = datetime.strptime(video["published_at"], "%Y-%m-%dT%H:%M:%SZ")
        if published_date.month == 2:  # Check if the video was published in February
            day = published_date.day
            views = int(video["stats"]["viewCount"])
            if day in views_by_day:
                views_by_day[day] += views
            else:
                views_by_day[day] = views
    return views_by_day

# Extract views for February from each JSON file
file_paths = ['sourav_joshi_data.json', 'flying_beast_data.json', 'mumbiker_nikhil_data.json']
views_data = [extract_views_for_february(file_path) for file_path in file_paths]

# Plot views for each file
days = range(1, 30)  # February has 28 days
plt.figure(figsize=(10, 6))
names = ['Sourav Joshi', 'Flying Beast', 'Mumbiker Nikhil']
for i, views_by_day in enumerate(views_data):
    views = [views_by_day.get(day, 0) for day in days]
    plt.plot(days, views, marker='o', label=names[i])

plt.title('Comparison of Views in February')
plt.xlabel('Day')
plt.ylabel('Views')
plt.xticks(days)
plt.grid(True)
plt.legend()
plt.show()
```

Comparing number of uploads in the month of February 2024

```python
import json
from datetime import datetime

# Function to count the number of uploads in February from a JSON file
def count_february_uploads(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    february_uploads = sum(1 for video in data if is_published_in_february(video))
    return february_uploads

# Function to check if a video is published in February
def is_published_in_february(video):
    published_date = datetime.strptime(video["published_at"], "%Y-%m-%dT%H:%M:%SZ")
    return published_date.month == 2

# Count the number of uploads in February from each JSON file
file_paths = ['sourav_joshi_data.json', 'flying_beast_data.json', 'mumbiker_nikhil_data.json']
february_upload_counts = [count_february_uploads(file_path) for file_path in file_paths]
names = ['Sourav Joshi', 'Flying Beast', 'Mumbiker Nikhil']

# Plot the number of uploads for each file
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(range(len(file_paths)), february_upload_counts, color='skyblue')
plt.xlabel('File')
plt.ylabel('Number of Uploads in February')
plt.title('Number of Uploads in February Comparison')
plt.xticks(range(len(file_paths)), [name for name in names])
plt.grid(axis='y')
plt.show()
```

Sourav Joshi

```python
import json
from datetime import datetime
import matplotlib.pyplot as plt

# Read JSON data from file
with open('sourav_joshi_data.json', 'r') as file:
    data = json.load(file)

# Extract likes and published dates for videos published in February
likes_by_day = {}
for video in data:
    published_date = datetime.strptime(video["published_at"], "%Y-%m-%dT%H:%M:%SZ")
    if published_date.month == 2:
        day = published_date.day
        likes = int(video["stats"]["likeCount"])
        if day in likes_by_day:
            likes_by_day[day] += likes
        else:
            likes_by_day[day] = likes

# Plot the likes for each day of February
days = list(likes_by_day.keys())
likes = list(likes_by_day.values())

plt.plot(days, likes, marker='o', linestyle='-', color='b')
plt.title('Likes for each day of February')
plt.xlabel('Day')
plt.ylabel('Likes')
plt.grid(True)
plt.show()
```

```python
import json
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Read JSON data from file
with open('sourav_joshi_data.json', 'r') as file:
    data = json.load(file)

# Extract titles from the data
titles = [video["title"] for video in data]

# Join all titles into a single string
all_titles = ' '.join(titles)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_titles)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud from Video Titles')
plt.axis('off')
plt.show()
```
